```python
import os
import pandas as pd
import chardet

file_paths = [
    r"C:\Users\User\Desktop\The Electricity Price Forecasting Challenge\Cleaned PowerCast Datasets\cleaned_Frequency_Containment_Reserve_202301010000_202503050000_Quarterhour.csv",
    r"C:\Users\User\Desktop\The Electricity Price Forecasting Challenge\Cleaned PowerCast Datasets\cleaned_Generation_Forecast_Intraday_202301010000_202503050000_Quarterhour.csv",
    r"C:\Users\User\Desktop\The Electricity Price Forecasting Challenge\Cleaned PowerCast Datasets\cleaned_Imported_balancing_services_202301010000_202503050000_Quarterhour.csv",
    r"C:\Users\User\Desktop\The Electricity Price Forecasting Challenge\Cleaned PowerCast Datasets\cleaned_Installed_generation_capacity_202301010000_202503050000_Year.csv",
    r"C:\Users\User\Desktop\The Electricity Price Forecasting Challenge\Cleaned PowerCast Datasets\cleaned_Manual_Frequency_Restoration_Reserve_202301010000_202503050000_Quarterhour.csv",
    r"C:\Users\User\Desktop\The Electricity Price Forecasting Challenge\Cleaned PowerCast Datasets\cleaned_Scheduled_commercial_exchanges_202301010000_202503050000_Quarterhour.csv",
    r"C:\Users\User\Desktop\The Electricity Price Forecasting Challenge\Cleaned PowerCast Datasets\cleaned_Actual_consumption_202301010000_202503050000_Quarterhour.csv",
    r"C:\Users\User\Desktop\The Electricity Price Forecasting Challenge\Cleaned PowerCast Datasets\cleaned_Actual_generation_202301010000_202503050000_Quarterhour.csv",
    r"C:\Users\User\Desktop\The Electricity Price Forecasting Challenge\Cleaned PowerCast Datasets\cleaned_Automatic_Frequency_Restoration_Reserve_202301010000_202503050000_Quarterhour.csv",
    r"C:\Users\User\Desktop\The Electricity Price Forecasting Challenge\Cleaned PowerCast Datasets\cleaned_Balancing_energy_202301010000_202503050000_Quarterhour_Month.csv",
    r"C:\Users\User\Desktop\The Electricity Price Forecasting Challenge\Cleaned PowerCast Datasets\cleaned_Costs_of_TSOs__without_costs_of_DSOs__202301010000_202503050000_Month.csv",
    r"C:\Users\User\Desktop\The Electricity Price Forecasting Challenge\Cleaned PowerCast Datasets\cleaned_Cross-border_physical_flows_202301010000_202503050000_Quarterhour.csv",
    r"C:\Users\User\Desktop\The Electricity Price Forecasting Challenge\Cleaned PowerCast Datasets\cleaned_Day-ahead_prices_202301010000_202503050000_Hour.csv",
    r"C:\Users\User\Desktop\The Electricity Price Forecasting Challenge\Cleaned PowerCast Datasets\cleaned_Exported_balancing_services_202301010000_202503050000_Quarterhour.csv",
    r"C:\Users\User\Desktop\The Electricity Price Forecasting Challenge\Cleaned PowerCast Datasets\cleaned_Forecasted_consumption_202301010000_202503050000_Quarterhour.csv",
    r"C:\Users\User\Desktop\The Electricity Price Forecasting Challenge\Cleaned PowerCast Datasets\cleaned_Forecasted_generation_Day-Ahead_202301010000_202503050000_Hour_Quarterhour.csv",
]

output_folder = r"C:\Users\User\Desktop\The Electricity Price Forecasting Challenge\Cleaned_Fixed_Datasets"
os.makedirs(output_folder, exist_ok=True)

def detect_encoding(file_path):
    with open(file_path, "rb") as f:
        result = chardet.detect(f.read(100000))  
        return result["encoding"]

def clean_data(df):
    missing_percentage = df.isnull().sum() / len(df)
    df = df.drop(columns=missing_percentage[missing_percentage >= 0.90].index)

    for col in df.columns:
        if df[col].dtype == "object" and col not in ["Start date", "End date"]:
            df[col] = df[col].astype(str).str.replace(",", "").str.replace(" ", "").str.replace("\xa0", "")
            df[col] = pd.to_numeric(df[col], errors="coerce")

    for col in df.columns:
        if col not in ["Start date", "End date"]:  
            if df[col].isnull().sum() > 0:
                missing_ratio = df[col].isnull().sum() / len(df[col])
                non_nan_values = df[col].dropna()

                if missing_ratio < 0.05:
                    df[col] = df[col].interpolate(method='linear', limit_direction='both')
                elif missing_ratio < 0.20:
                    df[col] = df[col].interpolate(method='quadratic', limit_direction='both')
                elif missing_ratio < 0.50:
                    df[col] = df[col].fillna(method='ffill').fillna(method='bfill')
                elif not non_nan_values.empty:  
                    df[col] = df[col].fillna(non_nan_values.median())
                else:
                    df[col] = df[col].fillna(0)  

    df = df.drop_duplicates()

    for col in df.select_dtypes(include=['float64']).columns:
        df[col] = df[col].astype('float32')

    for col in df.select_dtypes(include=['int64']).columns:
        df[col] = df[col].astype('int32')

    return df

for file in file_paths:
    try:
        detected_encoding = detect_encoding(file)
        print(f"📌 Detected encoding for {os.path.basename(file)}: {detected_encoding}")

        df = pd.read_csv(file, encoding=detected_encoding, low_memory=False)

        df = clean_data(df)

        output_file = os.path.join(output_folder, os.path.basename(file))
        df.to_csv(output_file, index=False, encoding=detected_encoding)

        print(f"✅ Cleaned file saved: {output_file}")

    except Exception as e:
        print(f"❌ Error processing file {file}: {e}")

print("🚀 Data cleaning and standardization process completed!")
```

```python
import pandas as pd
import numpy as np
import difflib

file_path = r"C:\Users\User\Desktop\The Electricity Price Forecasting Challenge\Cleaned_Fixed_Datasets\cleaned_Day-ahead_prices_202301010000_202503050000_Hour.csv"

try:
    df = pd.read_csv(file_path, encoding="utf-8")
except UnicodeDecodeError:
    df = pd.read_csv(file_path, encoding="ISO-8859-1")

target_column = "Northern Italy [€/MWh] Original resolutions"
best_match = difflib.get_close_matches(target_column, df.columns, n=1, cutoff=0.6)

if not best_match:
    raise ValueError(f"❌ Column '{target_column}' not found. Check column names: {df.columns.tolist()}")
else:
    column_name = best_match[0]  

missing_count = df[column_name].isna().sum()
total_rows = len(df)

if missing_count == 0:
    print(f"✅ No missing values in {column_name}. No action required.")
else:
    last_missing_index = df[column_name].last_valid_index()
    
    if last_missing_index is not None and last_missing_index < total_rows - missing_count:
        if df[column_name].diff().abs().median() < df[column_name].std() / 2:
            df[column_name] = df[column_name].interpolate(method='linear')
            method_used = "Linear Interpolation"
        else:
            df[column_name].fillna(df[column_name].median(), inplace=True)
            method_used = "Median Imputation"
    else:
        df[column_name].fillna(method='ffill', inplace=True)
        method_used = "Forward Fill"

    print(f"✅ Applied method for filling missing values: {method_used}")
    print(f"🔍 Total missing values before: {missing_count}, after: {df[column_name].isna().sum()}")

df.to_csv(file_path, index=False, encoding="utf-8")
print(f"📂 Changes have been saved directly to the original file: {file_path}")
```

```python
import pandas as pd
import os
import glob
import chardet

file_paths = glob.glob(r"C:\Users\User\Desktop\The Electricity Price Forecasting Challenge\Cleaned_Fixed_Datasets\*_Quarterhour*.csv")

prefix_mapping = {
    "Actual_consumption": "ACT_CONS_",
    "Actual_generation": "ACT_GEN_",
    "Automatic_Frequency_Restoration_Reserve": "RES_AFRR_",
    "Balancing_energy": "BAL_ENERGY_",
    "Cross-border_physical_flows": "CB_FLOW_",
    "Day-ahead_prices": "PRICE_DA_",
    "Exported_balancing_services": "BAL_EXP_",
    "Forecasted_consumption": "FCST_CONS_",
    "Forecasted_generation_Day-Ahead": "FCST_GEN_DA_",
    "Frequency_Containment_Reserve": "RES_FCR_",
    "Generation_Forecast_Intraday": "FCST_GEN_INTRADAY_",
    "Imported_balancing_services": "BAL_IMP_",
    "Manual_Frequency_Restoration_Reserve": "RES_MFRR_",
    "Scheduled_commercial_exchanges": "CB_SCHED_"
}

merged_df = None

for file in file_paths:
    filename = os.path.basename(file).replace("cleaned_", "").split("_2023")[0]  
    prefix = prefix_mapping.get(filename, "UNKNOWN_")  

    print(f"🔍 Procesăm fișierul: {filename}")

    encoding = "utf-8"
    try:
        with open(file, 'rb') as f:
            rawdata = f.read(100000)
        encoding = chardet.detect(rawdata)['encoding']
    except:
        print(f"⚠️ Nu am putut detecta encoding-ul pentru {file}. Se folosește UTF-8 implicit.")

    df = pd.read_csv(file, encoding=encoding)

    rename_dict = {col: f"{prefix}{col}" for col in df.columns if col not in ["Start date", "End date"]}
    df.rename(columns=rename_dict, inplace=True)

    df['Start date'] = pd.to_datetime(df['Start date'], errors='coerce')
    df['End date'] = pd.to_datetime(df['End date'], errors='coerce')

    if merged_df is None:
        merged_df = df
    else:
        merged_df = pd.merge(merged_df, df, on=['Start date', 'End date'], how='outer')

output_path = r"C:\Users\User\Desktop\The Electricity Price Forecasting Challenge\Cleaned_Fixed_Datasets\merged_Quarterhour.csv"
merged_df.to_csv(output_path, index=False, encoding='utf-8')

print(f"\n✅ Merge finalizat! Fișierul a fost salvat la: {output_path}")
```
